3.4.1 Решите систему уравнений:  
  y = x^2 – 1  
  exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 4.5, 10001)
plt.grid(True)
plt.plot(x, x*x - 1)
plt.plot(x, 1 - (1 - np.exp(x))/x);
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x*x - y - 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 5))
x2, y2 = fsolve(equations, (2, 5))
x3, y3 = fsolve(equations, (4, 15))

plt.grid(True)
plt.plot(x, x*x - 1)
plt.plot(x, 1 - (1 - np.exp(x))/x);

plt.scatter(x1, y1, color='red', zorder=3)
plt.scatter(x2, y2, color='green', zorder=3)
plt.scatter(x3, y3, color='blue', zorder=3)
plt.show()

print(f'x1 = {x1:6.4}  y1 = {y1:6.4}')
print(f'x2 = {x2:6.4}  y2 = {y2:6.4}')
print(f'x3 = {x3:6.4}  y3 = {y3:6.4}')
```
